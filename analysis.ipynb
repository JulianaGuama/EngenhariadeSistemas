## Qual a carga horária do curso por período?

```
df.select("Periodo", "Ch")\
  .groupby("Periodo")\
  .sum()\
  .withColumnRenamed("sum(Ch)", "Ch")\
  .display()
```

## Quantas disciplinas da universidade fazem parte do meu dia a dia?

```
# ponto de vista das disciplinas
df.groupby("Uso")\
  .count()\
  .display()
```

## Quanto da carga horária do curso faz parte do meu dia a dia?

```
# ponto de vista da carga horária
import pyspark.sql.functions as F
from pyspark.sql import Window

window = Window.rowsBetween(Window.unboundedPreceding,Window.unboundedFollowing)

df.groupBy("Uso")\
  .agg(F.sum("Ch").alias("SumCh"))\
  .withColumn("Total", F.sum(F.col("SumCh")).over(window))\
  .withColumn("Percent", F.col("SumCh")*100/F.col("total"))\
  .select("Uso", "Percent")\
  .display()
```

```
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

STOP_WORDS = stopwords.words('portuguese')
```

```
import pyspark.sql.types as types
import string
import re


def to_lower(text: str) -> str:
  """Lower text."""
  return text.lower()


def punctuation_removal(text: str) -> str:
  """Return text with '' instead punctuations."""
  return ''.join(ch for ch in text if ch not in string.punctuation)


def stop_words_removal(text: str) -> str:
  """Return text withou stop words."""
  pattern = re.compile(r"\b(" + r"|".join(STOP_WORDS) + r")\b\s*")
  return re.sub(pattern, "", text)


udf_to_lower = udf(lambda x: to_lower(x), types.StringType())
udf_punctuation_removal = udf(lambda x: punctuation_removal(x), types.StringType())
udf_stop_words_removal = udf(lambda x: stop_words_removal(x), types.StringType())
```

```
df = df.withColumn("tlower", udf_to_lower("Ementa"))\
       .withColumn("tpunctuation", udf_punctuation_removal("tlower"))\
       .withColumn("treatedEmenta", udf_stop_words_removal("tpunctuation"))\
       .select("Periodo", "Disciplina", "Ch", "Uso", "treatedEmenta", "Ementa")

df.display()
```

## Analisando ementas das disciplinas

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# gerando wordcloud
wordcloud = WordCloud(background_color="white",
                      width=1000,
                      height=1000,
                      max_words=2000,
                      max_font_size=200,
                      min_font_size=1)
```

### Ementas das disciplinas que fazem parte do meu dia a dia

```
# selecionando os dados
rows = df.filter("Uso = 'diario'")\
         .select("treatedEmenta")\
         .collect()

ementas_diarias = " ".join([row["treatedEmenta"] for row in rows])

# gerando wordcloud
wordcloud.generate(ementas_diarias)

# plot
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
```
